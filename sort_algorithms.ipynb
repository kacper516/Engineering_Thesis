```python
def load_data(file_path):
    return pd.read_csv(file_path, header=None, names=["Size", "Time"])
```

```python
def plot_sorting_algorithms_comparison_on_os_in_different_lang(algorithm, system):
    plt.figure(figsize=(10, 6))

    # Wszystkie systemy i ich odpowiednie pliki dla sortowania
    os_files = {
        "MacOS Sonoma": {
            "C++": files_macos_sonoma_cpp,
            "Rust": files_macos_sonoma_rust,
            "Go": files_macos_sonoma_go,
            "Python": files_macos_sonoma_python,
            "JavaScript JIT": files_macos_sonoma_js_jit,
            "JavaScript less JIT": files_macos_sonoma_js_less_jit,
        },
        "Ubuntu": {
            "C++": files_ubuntu_cpp,
            "Rust": files_ubuntu_rust,
            "Go": files_ubuntu_go,
            "Python": files_ubuntu_python,
            "JavaScript JIT": files_ubuntu_js_jit,
            "JavaScript less JIT": files_ubuntu_js_less_jit,
        },
        "EndeavourOS": {
            "C++": files_endeavouros_cpp,
            "Rust": files_endeavouros_rust,
            "Go": files_endeavouros_go,
            "Python": files_endeavouros_python,
            "JavaScript JIT": files_endeavouros_js_jit,
            "JavaScript less JIT": files_endeavouros_js_less_jit,
        },
        "Fedora": {
            "C++": files_fedora_cpp,
            "Rust": files_fedora_rust,
            "Go": files_fedora_go,
            "Python": files_fedora_python,
            "JavaScript JIT": files_fedora_js_jit,
            "JavaScript less JIT": files_fedora_js_less_jit,
        },
        "Windows11": {
            "C++": files_windows11_cpp,
            "Rust": files_windows11_rust,
            "Go": files_windows11_go,
            "Python": files_windows11_python,
            "JavaScript JIT": files_windows11_js_jit,
            "JavaScript less JIT": files_windows11_js_less_jit,
        },
    }

    # Sprawdzamy, czy wybrany system istnieje w słowniku
    if system not in os_files:
        raise ValueError(f"System {system} is not available.")

    # Wybieramy pliki dla danego systemu
    os_lang_files = os_files[system]

    # Słowniki do przechowywania czasów
    total_times = {}
    times_low = {}
    times_mid = {}
    times_high = {}

    for os_lang, files in os_lang_files.items():
        data = load_data(files[algorithm])
        plt.plot(data["Size"], data["Time"], label=f"{system} {os_lang}", marker="o")

        # Sumowanie czasów dla pełnych danych testowych
        total_time = sum(data["Time"])
        total_times[os_lang] = total_time

        # Sumowanie czasów dla przedziału 10 do 10^2 elementów
        times_low[os_lang] = sum(
            data["Time"][(data["Size"] >= 10) & (data["Size"] <= 10**2)]
        )

        # Czas wykonania dla 10^3 elementów
        times_mid[os_lang] = sum(data["Time"][(data["Size"] == 10**3)])

        # Sumowanie czasów dla przedziału 10^4 do 10^5 elementów
        times_high[os_lang] = sum(
            data["Time"][(data["Size"] >= 10**4) & (data["Size"] <= 10**5)]
        )

    # Konfiguracja osi i tytułów
    plt.xlabel("Array size")
    plt.ylabel("Time execution (ms)")
    title = f"Execution Time of {algorithm} on {system} Across Programming Languages"
    plt.title(title)
    plt.legend(loc="lower center", bbox_to_anchor=(0.8, -0.3), ncol=2)
    plt.grid(True)
    plt.xscale("log")  # Skala logarytmiczna na osi X
    plt.yscale("log")  # Skala logarytmiczna na osi Y

    # Wyświetlamy wykres
    plt.show()

    # Podsumowanie
    print(f"\nSummary of {algorithm} on {system} across programming languages:\n")

    # Całkowity czas wykonania
    max_lang_total = max(total_times, key=total_times.get)
    print(
        f"Language with highest total execution time: {max_lang_total} ({total_times[max_lang_total]:.2f} ms)"
    )
    min_total_lang = min(total_times, key=total_times.get)
    print(
        f"Language with lowest total execution time: {min_total_lang} ({total_times[min_total_lang]:.2f} ms)"
    )

    # Największy i najmniejszy czas wykonania dla przedziału 10 do 10^2 elementów
    max_lang_low = max(times_low, key=times_low.get)
    min_lang_low = min(times_low, key=times_low.get)
    print(
        f"Largest time for size 10 to 10^2: {max_lang_low} ({times_low[max_lang_low]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10 to 10^2: {min_lang_low} ({times_low[min_lang_low]:.2f} ms)"
    )

    # Czas wykonania dla 10^3 elementów
    max_lang_mid = max(times_mid, key=times_mid.get)
    min_lang_mid = min(times_mid, key=times_mid.get)
    print(
        f"Largest time for size 10^3: {max_lang_mid} ({times_mid[max_lang_mid]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10^3: {min_lang_mid} ({times_mid[min_lang_mid]:.2f} ms)"
    )

    # Największy i najmniejszy czas wykonania dla przedziału 10^4 do 10^5 elementów
    max_lang_high = max(times_high, key=times_high.get)
    min_lang_high = min(times_high, key=times_high.get)
    print(
        f"Largest time for size 10^4 to 10^5: {max_lang_high} ({times_high[max_lang_high]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10^4 to 10^5: {min_lang_high} ({times_high[min_lang_high]:.2f} ms)"
    )
```

```python
def plot_lang_comparison_on_different_os(lang, algorithm):
    plt.figure(figsize=(10, 6))

    # Wszystkie systemy i ich odpowiednie pliki dla sortowania
    os_files = {
        "MacOS Sonoma": {
            "C++": files_macos_sonoma_cpp,
            "Rust": files_macos_sonoma_rust,
            "Go": files_macos_sonoma_go,
            "Python": files_macos_sonoma_python,
            "JavaScript JIT": files_macos_sonoma_js_jit,
            "JavaScript less JIT": files_macos_sonoma_js_less_jit,
        },
        "Ubuntu": {
            "C++": files_ubuntu_cpp,
            "Rust": files_ubuntu_rust,
            "Go": files_ubuntu_go,
            "Python": files_ubuntu_python,
            "JavaScript JIT": files_ubuntu_js_jit,
            "JavaScript less JIT": files_ubuntu_js_less_jit,
        },
        "EndeavourOS": {
            "C++": files_endeavouros_cpp,
            "Rust": files_endeavouros_rust,
            "Go": files_endeavouros_go,
            "Python": files_endeavouros_python,
            "JavaScript JIT": files_endeavouros_js_jit,
            "JavaScript less JIT": files_endeavouros_js_less_jit,
        },
        "Fedora": {
            "C++": files_fedora_cpp,
            "Rust": files_fedora_rust,
            "Go": files_fedora_go,
            "Python": files_fedora_python,
            "JavaScript JIT": files_fedora_js_jit,
            "JavaScript less JIT": files_fedora_js_less_jit,
        },
        "Windows11": {
            "C++": files_windows11_cpp,
            "Rust": files_windows11_rust,
            "Go": files_windows11_go,
            "Python": files_windows11_python,
            "JavaScript JIT": files_windows11_js_jit,
            "JavaScript less JIT": files_windows11_js_less_jit,
        },
    }

    # Sprawdzamy, czy wybrany język istnieje dla wszystkich systemów
    systems = os_files.keys()
    lang_not_found = []

    for system in systems:
        if lang not in os_files[system]:
            lang_not_found.append(system)

    if lang_not_found:
        raise ValueError(
            f"Language {lang} is not available for systems: {', '.join(lang_not_found)}."
        )

    # Słowniki do przechowywania czasów dla przedziałów
    total_times = {}
    times_low = {}
    times_mid = {}
    times_high = {}

    # Wybieramy pliki dla danego języka na różnych systemach
    for system in systems:
        files = os_files[system][lang]
        data = load_data(files[algorithm])
        plt.plot(data["Size"], data["Time"], label=f"{system} {lang}", marker="o")

        # Sumowanie czasów dla różnych przedziałów
        total_times[system] = sum(data["Time"])
        times_low[system] = sum(
            data["Time"][(data["Size"] >= 10) & (data["Size"] <= 10**2)]
        )
        times_mid[system] = sum(data["Time"][(data["Size"] == 10**3)])
        times_high[system] = sum(
            data["Time"][(data["Size"] >= 10**4) & (data["Size"] <= 10**5)]
        )

    # Konfiguracja osi i tytułów
    plt.xlabel("Array size")
    plt.ylabel("Time execution (ms)")
    title = f"Execution Time of {algorithm} in {lang} Across Different OS"
    plt.title(title)
    plt.legend(loc="lower center", bbox_to_anchor=(0.8, -0.3), ncol=2)
    plt.grid(True)
    plt.xscale("log")  # Skala logarytmiczna na osi X
    plt.yscale("log")  # Skala logarytmiczna na osi Y

    # Wyświetlamy wykres
    plt.show()

    # Podsumowanie wyników dla różnych systemów
    print(f"\nSummary of {algorithm} in {lang} across different OS:\n")

    # Całkowity czas wykonania
    max_total_time = max(total_times, key=total_times.get)
    print(
        f"OS with highest total execution time: {max_total_time} ({total_times[max_total_time]:.2f} ms)"
    )
    min_total_time = min(total_times,key=total_times.get)
    print(
        f"OS with lowest total execution time: {min_total_time} ({total_times[min_total_time]:.2f} ms)"
    )

    # Największy i najmniejszy czas wykonania dla przedziału 10 do 10^2 elementów
    max_lang_low = max(times_low, key=times_low.get)
    min_lang_low = min(times_low, key=times_low.get)
    print(
        f"Largest time for size 10 to 10^2: {max_lang_low} ({times_low[max_lang_low]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10 to 10^2: {min_lang_low} ({times_low[min_lang_low]:.2f} ms)"
    )

    # Czas wykonania dla 10^3 elementów
    max_lang_mid = max(times_mid, key=times_mid.get)
    min_lang_mid = min(times_mid, key=times_mid.get)
    print(
        f"Largest time for size 10^3: {max_lang_mid} ({times_mid[max_lang_mid]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10^3: {min_lang_mid} ({times_mid[min_lang_mid]:.2f} ms)"
    )

    # Największy i najmniejszy czas wykonania dla przedziału 10^4 do 10^5 elementów
    max_lang_high = max(times_high, key=times_high.get)
    min_lang_high = min(times_high, key=times_high.get)
    print(
        f"Largest time for size 10^4 to 10^5: {max_lang_high} ({times_high[max_lang_high]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10^4 to 10^5: {min_lang_high} ({times_high[min_lang_high]:.2f} ms)"
    )
```
