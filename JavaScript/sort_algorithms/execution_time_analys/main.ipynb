```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Paths for files
sort_time_jit_windows11 = "../sort_execution_time/windows11_jit"
files_jit_windows11 = {
    "Heap Sort": f"{sort_time_jit_windows11}/heap_sort.csv",
    "Quick Sort": f"{sort_time_jit_windows11}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_jit_windows11}/bubble_sort.csv",
}

sort_time_less_jit_windows11 = "../sort_execution_time/windows11_without_jit"
files_less_jit_windows11 = {
    "Heap Sort": f"{sort_time_less_jit_windows11}/heap_sort.csv",
    "Quick Sort": f"{sort_time_less_jit_windows11}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_less_jit_windows11}/bubble_sort.csv",
}

# Paths for files
sort_time_jit_macos_sonoma = "../sort_execution_time/macos_sonoma_jit"
files_jit_macos_sonoma = {
    "Heap Sort": f"{sort_time_jit_macos_sonoma}/heap_sort.csv",
    "Quick Sort": f"{sort_time_jit_macos_sonoma}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_jit_macos_sonoma}/bubble_sort.csv",
}

sort_time_less_jit_macos_sonoma = "../sort_execution_time/macos_sonoma_without_jit"
files_less_jit_macos_sonoma = {
    "Heap Sort": f"{sort_time_less_jit_macos_sonoma}/heap_sort.csv",
    "Quick Sort": f"{sort_time_less_jit_macos_sonoma}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_less_jit_macos_sonoma}/bubble_sort.csv",
}

# Paths for files
sort_time_jit_ubuntu = "../sort_execution_time/ubuntu_jit"
files_jit_ubuntu = {
    "Heap Sort": f"{sort_time_jit_ubuntu}/heap_sort.csv",
    "Quick Sort": f"{sort_time_jit_ubuntu}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_jit_ubuntu}/bubble_sort.csv",
}

sort_time_less_jit_ubuntu = "../sort_execution_time/ubuntu_without_jit"
files_less_jit_ubuntu = {
    "Heap Sort": f"{sort_time_less_jit_ubuntu}/heap_sort.csv",
    "Quick Sort": f"{sort_time_less_jit_ubuntu}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_less_jit_ubuntu}/bubble_sort.csv",
}

# Paths for files
sort_time_jit_endeavouros = "../sort_execution_time/endeavouros_jit"
files_jit_endeavouros = {
    "Heap Sort": f"{sort_time_jit_endeavouros}/heap_sort.csv",
    "Quick Sort": f"{sort_time_jit_endeavouros}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_jit_endeavouros}/bubble_sort.csv",
}

sort_time_less_jit_endeavouros = "../sort_execution_time/endeavouros_without_jit"
files_less_jit_endeavouros = {
    "Heap Sort": f"{sort_time_less_jit_endeavouros}/heap_sort.csv",
    "Quick Sort": f"{sort_time_less_jit_endeavouros}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_less_jit_endeavouros}/bubble_sort.csv",
}

# Paths for files
sort_time_jit_fedora = "../sort_execution_time/fedora_jit"
files_jit_fedora = {
    "Heap Sort": f"{sort_time_jit_fedora}/heap_sort.csv",
    "Quick Sort": f"{sort_time_jit_fedora}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_jit_fedora}/bubble_sort.csv",
}

sort_time_less_jit_fedora = "../sort_execution_time/fedora_without_jit"
files_less_jit_fedora = {
    "Heap Sort": f"{sort_time_less_jit_fedora}/heap_sort.csv",
    "Quick Sort": f"{sort_time_less_jit_fedora}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_less_jit_fedora}/bubble_sort.csv",
}
```

```python
def load_data(file_path):
    return pd.read_csv(file_path, header=None, names=["Size", "Time"])
```

```python
def plot_comparison(data_dict, operating_system="Windows", if_jit_less=False):
    """Rysuje wykres porównujący czasy wykonania algorytmów."""
    plt.figure(figsize=(10, 6))

    # Przechodzimy po danych i rysujemy wykres dla każdego algorytmu
    for label, data in data_dict.items():
        plt.plot(data["Size"], data["Time"], label=label, marker="o")

    # Konfiguracja osi i tytułów
    plt.xlabel("Array size")
    plt.ylabel("Time execution (ms)")

    title = f"Execution Time of Sort Algorithms in JavaScript {'without' if if_jit_less else 'with'} JIT on {operating_system}"

    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.xscale(
        "log"
    )  # Skala logarytmiczna na osi X, aby uwzględnić duże rozmiary danych
    plt.yscale("log")  # Skala logarytmiczna na osi Y, dla porównania różnic czasów

    # Wyświetlamy wykres
    plt.show()
```

```python
def plot_algorithm_comparison_across_os_js_jit(algorithm):
    """Rysuje wykres porównujący czasy wykonania algorytmu dla różnych systemów operacyjnych (JS JIT)."""
    plt.figure(figsize=(10, 6))

    os_files = {
        "Windows11": files_jit_windows11,
        "MacOS Sonoma": files_jit_macos_sonoma,
        "Ubuntu": files_jit_ubuntu,
        "EndeavourOS": files_jit_endeavouros,
        "Fedora": files_jit_fedora,
    }

    for os_name, files in os_files.items():
        data = load_data(files[algorithm])
        plt.plot(data["Size"], data["Time"], label=os_name, marker="o")

    plt.xlabel("Array size")
    plt.ylabel("Time execution (ms)")
    title = f"Execution Time of {algorithm} in JavaScript JIT Across Operating Systems"
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.xscale("log")
    plt.yscale("log")
    plt.show()
```

```python
def plot_algorithm_comparison_across_os_js_no_jit(algorithm):
    """Rysuje wykres porównujący czasy wykonania algorytmu dla różnych systemów operacyjnych (JS bez JIT)."""
    plt.figure(figsize=(10, 6))

    os_files = {
        "Windows11": files_less_jit_windows11,
        "MacOS Sonoma": files_less_jit_macos_sonoma,
        "Ubuntu": files_less_jit_ubuntu,
        "EndeavourOS": files_less_jit_endeavouros,
        "Fedora": files_less_jit_fedora,
    }

    for os_name, files in os_files.items():
        data = load_data(files[algorithm])
        plt.plot(data["Size"], data["Time"], label=os_name, marker="o")

    plt.xlabel("Array size")
    plt.ylabel("Time execution (ms)")
    title = f"Execution Time of {algorithm} in JavaScript Without JIT Across Operating Systems"
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.xscale("log")
    plt.yscale("log")
    plt.show()
```

```python
def plot_algorithm_comparison_across_os_js_both(algorithm):
    """Rysuje wykres porównujący czasy wykonania algorytmu dla różnych systemów operacyjnych (JS JIT i bez JIT)."""
    plt.figure(figsize=(10, 6))

    os_files_jit = {
        "Windows11 JIT": files_jit_windows11,
        "MacOS Sonoma JIT": files_jit_macos_sonoma,
        "Ubuntu JIT": files_jit_ubuntu,
        "EndeavourOS JIT": files_jit_endeavouros,
        "Fedora JIT": files_jit_fedora,
    }

    os_files_no_jit = {
        "Windows11 No JIT": files_less_jit_windows11,
        "MacOS Sonoma No JIT": files_less_jit_macos_sonoma,
        "Ubuntu No JIT": files_less_jit_ubuntu,
        "EndeavourOS No JIT": files_less_jit_endeavouros,
        "Fedora No JIT": files_less_jit_fedora,
    }

    for os_name, files in os_files_jit.items():
        data = load_data(files[algorithm])
        plt.plot(data["Size"], data["Time"], label=os_name, marker="o")

    for os_name, files in os_files_no_jit.items():
        data = load_data(files[algorithm])
        plt.plot(data["Size"], data["Time"], label=os_name, marker="x")

    plt.xlabel("Array size")
    plt.ylabel("Time execution (ms)")
    title = f"Execution Time of {algorithm} in JavaScript JIT and Without JIT Across Operating Systems"
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.xscale("log")
    plt.yscale("log")
    plt.show()
```

```python
def compare_algorithms(operating_system="Windows11", if_jit_less=False):
    # Wczytujemy dane z plików CSV
    if operating_system == "Windows11" and not if_jit_less:
        files = files_jit_windows11
    elif operating_system == "Windows11" and if_jit_less:
        files = files_less_jit_windows11
    elif operating_system == "MacOS Sonoma" and not if_jit_less:
        files = files_jit_macos_sonoma
    elif operating_system == "MacOS Sonoma" and if_jit_less:
        files = files_less_jit_macos_sonoma
    elif operating_system == "Ubuntu" and not if_jit_less:
        files = files_jit_ubuntu
    elif operating_system == "Ubuntu" and if_jit_less:
        files = files_less_jit_ubuntu
    elif operating_system == "EndeavourOS" and not if_jit_less:
        files = files_jit_endeavouros
    elif operating_system == "EndeavourOS" and if_jit_less:
        files = files_less_jit_endeavouros
    elif operating_system == "Fedora" and not if_jit_less:
        files = files_jit_fedora
    elif operating_system == "Fedora" and if_jit_less:
        files = files_less_jit_fedora

    data_dict = {label: load_data(file_path) for label, file_path in files.items()}

    # Wywołujemy funkcję do rysowania wykresu
    plot_comparison(data_dict, operating_system, if_jit_less)
```
