```python
def load_data(file_path):
    return pd.read_csv(file_path, header=None, names=["Size", "Time"])
```

```python
def plot_algorithm_comparison_on_os_in_different_lang(algorithm, system):
    plt.figure(figsize=(10, 6))

    # Wszystkie systemy i ich odpowiednie pliki
    os_files = {
        "MacOS Sonoma": {
            "C++": files_macos_sonoma_cpp,
            "Rust": files_macos_sonoma_rust,
            "Go": files_macos_sonoma_go,
            "Python": files_macos_sonoma_python,
            "JavaScript JIT": files_macos_sonoma_js_jit,
            "JavaScript less JIT": files_macos_sonoma_js_less_jit,
        },
        "Ubuntu": {
            "C++": files_ubuntu_cpp,
            "Rust": files_ubuntu_rust,
            "Go": files_ubuntu_go,
            "Python": files_ubuntu_python,
            "JavaScript JIT": files_ubuntu_js_jit,
            "JavaScript less JIT": files_ubuntu_js_less_jit,
        },
        "EndeavourOS": {
            "C++": files_endeavouros_cpp,
            "Rust": files_endeavouros_rust,
            "Go": files_endeavouros_go,
            "Python": files_endeavouros_python,
            "JavaScript JIT": files_endeavouros_js_jit,
            "JavaScript less JIT": files_endeavouros_js_less_jit,
        },
        "Fedora": {
            "C++": files_fedora_cpp,
            "Rust": files_fedora_rust,
            "Go": files_fedora_go,
            "Python": files_fedora_python,
            "JavaScript JIT": files_fedora_js_jit,
            "JavaScript less JIT": files_fedora_js_less_jit,
        },
        "Windows11": {
            "C++": files_windows11_cpp,
            "Rust": files_windows11_rust,
            "Go": files_windows11_go,
            "Python": files_windows11_python,
            "JavaScript JIT": files_windows11_js_jit,
            "JavaScript less JIT": files_windows11_js_less_jit,
        },
    }

    # Sprawdzamy, czy wybrany system istnieje w słowniku
    if system not in os_files:
        raise ValueError(f"System {system} is not available.")

    # Wybieramy pliki dla danego systemu
    os_lang_files = os_files[system]

    total_times = {}
    max_times = {}
    min_times = {}
    times_10_3 = {}
    times_10_5 = {}
    times_10_8 = {}

    # Iteracja po plikach dla każdego języka
    for os_lang, files in os_lang_files.items():
        data = load_data(files[algorithm])
        plt.plot(data["Size"], data["Time"], label=f"{system} {os_lang}", marker="o")

        # Sumowanie czasów dla pełnych danych testowych
        total_time = sum(data["Time"])
        total_times[os_lang] = total_time

        # Największy i najmniejszy czas dla pełnego zakresu danych
        max_time = max(data["Time"])
        min_time = min(data["Time"])
        max_times[os_lang] = max_time
        min_times[os_lang] = min_time

        # Sumowanie czasów dla przedziału 10 elementów do 10^3 elementów
        times_10_3[os_lang] = sum(
            data["Time"][(data["Size"] >= 10) & (data["Size"] <= 10**3)]
        )

        # Sumowanie czasów dla przedziału 10^4 elementów do 10^5 elementów
        times_10_5[os_lang] = sum(
            data["Time"][(data["Size"] >= 10**4) & (data["Size"] <= 10**5)]
        )

        # Sumowanie czasów dla przedziału 10^6 elementów do 10^8 elementów
        times_10_8[os_lang] = sum(
            data["Time"][(data["Size"] >= 10**6) & (data["Size"] <= 10**8)]
        )

    # Konfiguracja osi i tytułów
    plt.xlabel("Array size")
    plt.ylabel("Time execution (ms)")
    title = f"Execution Time of {algorithm} on {system} Across Programming Languages"
    plt.title(title)
    plt.legend(loc="lower center", bbox_to_anchor=(0.9, -0.4), fontsize="small")
    plt.grid(True)
    plt.xscale("log")
    plt.yscale("log")

    # Wyświetlenie wykresu
    plt.show()

    # Podsumowanie
    print(f"\nSummary of {algorithm} on {system}:\n")

    # Całkowity czas wykonania
    max_lang_total = max(total_times, key=total_times.get)
    print(
        f"Language with highest total execution time: {max_lang_total} ({total_times[max_lang_total]:.2f} ms)"
    )
    min_lang_total = min(total_times, key=total_times.get)
    print(
        f"Language with lowest total execution time: {min_lang_total} ({total_times[min_lang_total]:.2f} ms)"
    )

    # Największy i najmniejszy czas wykonania dla 10 - 10^3 elementów
    max_lang_10_3 = max(times_10_3, key=times_10_3.get)
    min_lang_10_3 = min(times_10_3, key=times_10_3.get)
    print(
        f"Largest time for size 10 to 10^3: {max_lang_10_3} ({times_10_3[max_lang_10_3]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10 to 10^3: {min_lang_10_3} ({times_10_3[min_lang_10_3]:.2f} ms)"
    )

    # Największy i najmniejszy czas wykonania dla 10^4 - 10^5 elementów
    max_lang_10_5 = max(times_10_5, key=times_10_5.get)
    min_lang_10_5 = min(times_10_5, key=times_10_5.get)
    print(
        f"Largest time for size 10^4 to 10^5: {max_lang_10_5} ({times_10_5[max_lang_10_5]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10^4 to 10^5: {min_lang_10_5} ({times_10_5[min_lang_10_5]:.2f} ms)"
    )

    # Największy i najmniejszy czas wykonania dla 10^6 - 10^8 elementów
    max_lang_10_8 = max(times_10_8, key=times_10_8.get)
    min_lang_10_8 = min(times_10_8, key=times_10_8.get)
    print(
        f"Largest time for size 10^6 to 10^8: {max_lang_10_8} ({times_10_8[max_lang_10_8]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10^6 to 10^8: {min_lang_10_8} ({times_10_8[min_lang_10_8]:.2f} ms)"
    )
```

```python
def plot_lang_comparison_on_different_os(lang, algorithm):
    plt.figure(figsize=(10, 6))
    ...

    # Wszystkie systemy i ich odpowiednie pliki
    os_files = {
        "MacOS Sonoma": {
            "C++": files_macos_sonoma_cpp,
            "Rust": files_macos_sonoma_rust,
            "Go": files_macos_sonoma_go,
            "Python": files_macos_sonoma_python,
            "JavaScript JIT": files_macos_sonoma_js_jit,
            "JavaScript less JIT": files_macos_sonoma_js_less_jit,
        },
        "Ubuntu": {
            "C++": files_ubuntu_cpp,
            "Rust": files_ubuntu_rust,
            "Go": files_ubuntu_go,
            "Python": files_ubuntu_python,
            "JavaScript JIT": files_ubuntu_js_jit,
            "JavaScript less JIT": files_ubuntu_js_less_jit,
        },
        "EndeavourOS": {
            "C++": files_endeavouros_cpp,
            "Rust": files_endeavouros_rust,
            "Go": files_endeavouros_go,
            "Python": files_endeavouros_python,
            "JavaScript JIT": files_endeavouros_js_jit,
            "JavaScript less JIT": files_endeavouros_js_less_jit,
        },
        "Fedora": {
            "C++": files_fedora_cpp,
            "Rust": files_fedora_rust,
            "Go": files_fedora_go,
            "Python": files_fedora_python,
            "JavaScript JIT": files_fedora_js_jit,
            "JavaScript less JIT": files_fedora_js_less_jit,
        },
        "Windows11": {
            "C++": files_windows11_cpp,
            "Rust": files_windows11_rust,
            "Go": files_windows11_go,
            "Python": files_windows11_python,
            "JavaScript JIT": files_windows11_js_jit,
            "JavaScript less JIT": files_windows11_js_less_jit,
        },
    }

    # Sprawdzamy, czy wybrany język istnieje dla wszystkich systemów
    systems = os_files.keys()
    lang_not_found = []

    for system in systems:
        if lang not in os_files[system]:
            lang_not_found.append(system)

    if lang_not_found:
        raise ValueError(
            f"Language {lang} is not available for systems: {', '.join(lang_not_found)}."
        )

    # Słowniki do przechowywania czasów
    total_times = {}
    max_times = {}
    min_times = {}
    times_10_3 = {}
    times_10_5 = {}
    times_10_8 = {}

    # Wybieramy pliki dla danego języka na różnych systemach
    for system in systems:
        files = os_files[system][lang]
        data = load_data(files[algorithm])
        plt.plot(data["Size"], data["Time"], label=f"{system} {lang}", marker="o")

        # Sumowanie czasów dla pełnych danych testowych
        total_time = sum(data["Time"])
        total_times[system] = total_time

        # Największy i najmniejszy czas dla pełnego zakresu danych
        max_time = max(data["Time"])
        min_time = min(data["Time"])
        max_times[system] = max_time
        min_times[system] = min_time

        # Sumowanie czasów dla przedziału 10 elementów do 10^3 elementów
        times_10_3[system] = sum(
            data["Time"][(data["Size"] >= 10) & (data["Size"] <= 10**3)]
        )

        # Sumowanie czasów dla przedziału 10^4 elementów do 10^5 elementów
        times_10_5[system] = sum(
            data["Time"][(data["Size"] >= 10**4) & (data["Size"] <= 10**5)]
        )

        # Sumowanie czasów dla przedziału 10^6 elementów do 10^8 elementów
        times_10_8[system] = sum(
            data["Time"][(data["Size"] >= 10**6) & (data["Size"] <= 10**8)]
        )

    # Konfiguracja osi i tytułów
    plt.xlabel("Array size")
    plt.ylabel("Time execution (ms)")
    title = f"Execution Time of {algorithm} in {lang} Across Different OS"
    plt.title(title)
    plt.legend(loc="lower center", bbox_to_anchor=(0.9, -0.4), ncol=2, fontsize="small")
    plt.grid(True)
    plt.xscale("log")  # Skala logarytmiczna na osi X
    plt.yscale("log")  # Skala logarytmiczna na osi Y

    # Wyświetlamy wykres
    plt.show()

    # Podsumowanie
    print(f"\nSummary of {algorithm} in {lang} across different OS:\n")

    # Całkowity czas wykonania
    max_os_total = max(total_times, key=total_times.get)
    print(
        f"OS with highest total execution time: {max_os_total} ({total_times[max_os_total]:.2f} ms)"
    )
    min_os_total = min(total_times, key=total_times.get)
    print(
        f"OS with lowest total execution time: {min_os_total} ({total_times[min_os_total]:.2f} ms)"
    )

    # Największy i najmniejszy czas wykonania dla 10 - 10^3 elementów
    max_os_10_3 = max(times_10_3, key=times_10_3.get)
    min_os_10_3 = min(times_10_3, key=times_10_3.get)
    print(
        f"Largest time for size 10 to 10^3: {max_os_10_3} ({times_10_3[max_os_10_3]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10 to 10^3: {min_os_10_3} ({times_10_3[min_os_10_3]:.2f} ms)"
    )

    # Największy i najmniejszy czas wykonania dla 10^4 - 10^5 elementów
    max_os_10_5 = max(times_10_5, key=times_10_5.get)
    min_os_10_5 = min(times_10_5, key=times_10_5.get)
    print(
        f"Largest time for size 10^4 to 10^5: {max_os_10_5} ({times_10_5[max_os_10_5]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10^4 to 10^5: {min_os_10_5} ({times_10_5[min_os_10_5]:.2f} ms)"
    )

    # Największy i najmniejszy czas wykonania dla 10^6 - 10^8 elementów
    max_os_10_8 = max(times_10_8, key=times_10_8.get)
    min_os_10_8 = min(times_10_8, key=times_10_8.get)
    print(
        f"Largest time for size 10^6 to 10^8: {max_os_10_8} ({times_10_8[max_os_10_8]:.2f} ms)"
    )
    print(
        f"Smallest time for size 10^6 to 10^8: {min_os_10_8} ({times_10_8[min_os_10_8]:.2f} ms)"
    )
```
