```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sort_time_windows11 = "../sort_execution_time/windows11"
files_windows11 = {
    "Heap Sort": f"{sort_time_windows11}/heap_sort.csv",
    "Quick Sort": f"{sort_time_windows11}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_windows11}/bubble_sort.csv",
}

sort_time_macos_sonoma = "../sort_execution_time/macos_sonoma"
files_macos_sonoma = {
    "Heap Sort": f"{sort_time_macos_sonoma}/heap_sort.csv",
    "Quick Sort": f"{sort_time_macos_sonoma}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_macos_sonoma}/bubble_sort.csv",
}

sort_time_ubuntu = "../sort_execution_time/ubuntu"
files_ubuntu = {
    "Heap Sort": f"{sort_time_ubuntu}/heap_sort.csv",
    "Quick Sort": f"{sort_time_ubuntu}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_ubuntu}/bubble_sort.csv",
}

sort_time_endeavouros = "../sort_execution_time/endeavouros"
files_endeavouros = {
    "Heap Sort": f"{sort_time_endeavouros}/heap_sort.csv",
    "Quick Sort": f"{sort_time_endeavouros}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_endeavouros}/bubble_sort.csv",
}

sort_time_fedora = "../sort_execution_time/fedora"
files_fedora = {
    "Heap Sort": f"{sort_time_fedora}/heap_sort.csv",
    "Quick Sort": f"{sort_time_fedora}/quick_sort.csv",
    "Bubble Sort": f"{sort_time_fedora}/bubble_sort.csv",
}
```

```python
def load_data(file_path):
    return pd.read_csv(file_path, header=None, names=["Size", "Time"])
```

```python
def plot_comparison(data_dict, operating_system):
    """Rysuje wykres porównujący czasy wykonania algorytmów sortowania."""
    plt.figure(figsize=(10, 6))

    # Przechodzimy po danych i rysujemy wykres dla każdego algorytmu
    for label, data in data_dict.items():
        plt.plot(data["Size"], data["Time"], label=label, marker="o")

    # Konfiguracja osi i tytułów
    plt.xlabel("Array size")
    plt.ylabel("Time execution (ms)")
    title = "Execution Time of Sort Algorithms in Go on " + operating_system
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.xscale(
        "log"
    )  # Skala logarytmiczna na osi X, aby uwzględnić duże rozmiary danych
    plt.yscale("log")  # Skala logarytmiczna na osi Y, dla porównania różnic czasów

    # Wyświetlamy wykres
    plt.show()
```

```python
def plot_algorithm_comparison_across_os(algorithm):
    plt.figure(figsize=(10, 6))

    os_files = {
        "Windows11": files_windows11,
        "MacOS Sonoma": files_macos_sonoma,
        "Ubuntu": files_ubuntu,
        "EndeavourOS": files_endeavouros,
        "Fedora": files_fedora,
    }

    for os_name, files in os_files.items():
        data = load_data(files[algorithm])
        plt.plot(data["Size"], data["Time"], label=os_name, marker="o")

    # Konfiguracja osi i tytułów
    plt.xlabel("Array size")
    plt.ylabel("Time execution (ms)")
    title = f"Execution Time of {algorithm} in GO Across Operating Systems"
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.xscale(
        "log"
    )  # Skala logarytmiczna na osi X, aby uwzględnić duże rozmiary danych
    plt.yscale("log")  # Skala logarytmiczna na osi Y, dla porównania różnic czasów

    # Wyświetlamy wykres
    plt.show()
```

```python
def compare_algorithms(operating_system="Windows11"):
    # Wczytujemy dane z plików CSV

    files = files_windows11
    if operating_system == "MacOS Sonoma":
        files = files_macos_sonoma
    elif operating_system == "Ubuntu":
        files = files_ubuntu
    elif operating_system == "EndeavourOS":
        files = files_endeavouros
    elif operating_system == "Fedora":
        files = files_fedora

    data_dict = {label: load_data(file_path) for label, file_path in files.items()}

    # Wywołujemy funkcję do rysowania wykresu
    plot_comparison(data_dict, operating_system)
```

```python
compare_algorithms()
```

```python
compare_algorithms("MacOS Sonoma")
```

```python
compare_algorithms("Ubuntu")
```

```python
compare_algorithms("EndeavourOS")
```

```python
compare_algorithms("Fedora")
```

```python
plot_algorithm_comparison_across_os("Bubble Sort")
```

```python
plot_algorithm_comparison_across_os("Heap Sort")
```

```python
plot_algorithm_comparison_across_os("Quick Sort")
```
